```python
## Feature Engineering 

## Date_of_Journey ........

df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
## Performing the cleaning on Arrival_Time

df['arrival_hr']=df['Arrival_Time'].str.split(':').str[0]
df['arrival_min']=df['Arrival_Time'].str.split(':').str[1].str.split(' ').str[0]
```

```python
df.head(2)
```

```python
df['arrival_hr']=df['arrival_hr'].astype(int)
df['arrival_min']=df['arrival_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
## Performing Cleaning on the Dep_Time

df['dep_hr']=df['Dep_Time'].str.split(':').str[0]
df['dep_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['dep_hr']=df['dep_hr'].astype(int)
df['dep_min']=df['dep_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
## performing the cleaning on the duration 

df['Duration_hr']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.tail(2)
```

```python
df['Duration_min']=df['Duration_min'].fillna(0)
df.head()
```
