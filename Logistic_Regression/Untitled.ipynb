```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
X.shape
```

```python
## spliting the dataset in training and test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape,X_test.shape
```

## As we have  generated the dataset from the make_classification so we donot have to Standardized the data_set

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train,Y_train)
```

```python
y_pred=logistic.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
score=accuracy_score(Y_test,y_pred)
print(score)
print(classification_report(Y_test,y_pred))
print(confusion_matrix(Y_test,y_pred))
```

### Hyperparameter tunning and Cross_validation

In this we have to study two type of  Hyperparameter Tunning 
### 1. Grid Search Hyper Parameter

```python
model=LogisticRegression()
params={
    'penalty':['l1', 'l2', 'elasticnet'],
    'C':[100,10,1.0,0.1],
    'solver':['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}
```

```python
from sklearn.model_selection import  StratifiedKFold
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
score=accuracy_score(Y_test,y_pred)
print(score)
print(classification_report(Y_test,y_pred))
print(confusion_matrix(Y_test,y_pred))
```

```python
## this accuracy we get from the training data only while performing the cross validation , while the accuracy
## which we calculated above comes on testing againt test_data
grid.best_score_
```

### 2. Randomized Search Cv

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model=LogisticRegression()
random=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,n_jobs=-1,scoring='accuracy')
```

```python
random
```

```python
random.fit(X_train,Y_train)
```

```python
random.best_score_
```

```python
random.best_params_
```

```python
y_pred=random.predict(X_test)
```

```python
score=accuracy_score(Y_test,y_pred)
print(score)
print(classification_report(Y_test,y_pred))
print(confusion_matrix(Y_test,y_pred))
```

### IF we have Multi-Class output

```python
## n_informative is used to give how many features are related to the output field , and usefull for the classification 
X,y=make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=42)
```

```python
X.shape
```

```python
## let us split the dataset into train and test data
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.30,random_state=42)
X_train.shape ,X_test.shape , Y_train.shape
```

```python
multi=LogisticRegression(multi_class='ovr')
```

```python
multi.fit(X_train,Y_train)
```

```python
y_pred=multi.predict(X_test)
score=accuracy_score(Y_test,y_pred)
print(score)
print(classification_report(Y_test,y_pred))
print(confusion_matrix(Y_test,y_pred))
```

```python
grd=GridSearchCV(estimator=multi,param_grid=params,scoring='accuracy',n_jobs=-1,cv=cv)
```

```python
grd
```

```python
grd.fit(X_train,Y_train)
```

```python
grd.best_score_
```

```python
ran=RandomizedSearchCV(estimator=multi,param_distributions=params,cv=cv,scoring='accuracy',n_jobs=-1)
```

```python
ran.fit(X_train,Y_train)
```

```python
ran.best_score_
```
