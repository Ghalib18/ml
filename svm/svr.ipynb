```python
## we are going to used the dataset that where provided by the seaborn library
import seaborn as sns
import pandas as pd
df=sns.load_dataset('tips')
```

**we have to perform the EDA and Feature - engineering in the given dataset**

```python
df.isnull().sum()
## thus there is  no nullValue, so we can proceeds with futher featuring-engineering
```

```python
## we have 4 categorial feature which need to be encoded
df['time'].value_counts()
```

```python
df['day'].value_counts()
```

```python
## let us  find the dependent and independent dataset
x=df[['tip','sex','smoker','day','time','size']]
y=df['total_bill']
```

```python
## spliting the train and test datasets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## for the binary class we can easily used the lable_encoding and for day we can use OneHotEncoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l1=LabelEncoder()
l2=LabelEncoder()
l3=LabelEncoder()
```

```python
import warnings
warnings.filterwarnings('ignore')
X_train['time']=l1.fit_transform(X_train['time'])
X_train['sex']=l2.fit_transform(X_train['sex'])
X_train['smoker']=l3.fit_transform(X_train['smoker'])
```

```python
X_test['time']=l1.transform(X_test['time'])
X_test['sex']=l2.transform(X_test['sex'])
X_test['smoker']=l3.transform(X_test['smoker'])
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
## now its time to perform the OneHotencoding
```

```python
## OneHotencoding through column-transform
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])],
                                   remainder='passthrough')
```

```python

X_train=ct.fit_transform(X_train)
```

```python
X_test=ct.transform(X_test)
```

```python
X_train
```

```python
pd.DataFrame(X_train).info()
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(X_train,Y_train)
```

```python
pd.DataFrame(X_test).info()
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
y_pred=svr.predict(X_test)
score=r2_score(Y_test,y_pred)
mae=mean_absolute_error(Y_test,y_pred)
print(score)
print(mae)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid=GridSearchCV(SVR(),param_grid=param_grid,cv=5,verbose=3,refit=True)
```

```python
grid.fit(X_train,Y_train)
```
