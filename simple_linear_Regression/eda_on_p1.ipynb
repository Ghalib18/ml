## Data Cleaning

```python
df.dropna(inplace=True)
```

```python
## creating the new field which will categorised the the region.... 
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
```

```python
df['Region']=df['Region'].astype(int)
df.head()
```

```python
df.iloc[122]
```

```python
df.head(123)
```

```python
df=df.drop(123,axis=0)
```

```python

df.reset_index(drop=True,inplace=True)
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
## changing the type of the features in int -type
df[['day','month','year','Temperature','RH','Ws','Rain']]=df[['day','month','year','Temperature','RH','Ws','Rain']].astype(float).astype(int)
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
## we are going to save the clean_csv file
df.to_csv('Algerian_clean_data.csv',index=False)
```
